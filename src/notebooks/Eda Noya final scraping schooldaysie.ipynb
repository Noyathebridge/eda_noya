```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

```python
url_principal= "https://www.schooldays.ie"
```

Empezamos por las zonas


```python
url = 'https://www.schooldays.ie/articles/secondary-Schools-in-Ireland-by-County'
response = requests.get(url)
soup_url = bs(response.content, 'lxml')
response
```

Creamos una lista con todas las zonas

```python
span = soup_url.find('span', class_= 'dubSpan')
lista_a = span.find_all('a')


lista_zonas = []

for url in lista_a:
    if url_principal+url['href'] not in lista_zonas:
        lista_zonas.append(url_principal + url['href'])

lista_zonas
```

Ahora creamos una lista que nos saque todos los colegios de todas las zonas

```python
lista_colegios = []
for zona in lista_zonas:
    response = requests.get(zona)
    soup_zona = bs(response.content, 'lxml')
    #Accedo al contenedor donde se encuentra la información
    container = soup_zona.find('div', class_= 'SearchListingsView1')
    #Dentro del contenedor, busco las etiquetas tr
    trs = container.find_all('tr')
    for link in trs:
        try: #Si la etiqueta tr NO TIENE etiqueta "a", sigue buscando en las demás tr

        #Queremos el href, que es el trozo de url que cambia entre un colegio y otro
            url_secundaria = link.find('a')['href'] 

        #Vemos si las urls variables están seleccionadas corrrectamente
            print(url_secundaria) 

        #las metemos dentro de una lista 
            lista_colegios.append(url_principal+url_secundaria)
        #encadenamos la url principal o fija y la secundaria o variable

        except:
            continue
print(lista_colegios[0:30])
print(len(lista_colegios))
    
```
